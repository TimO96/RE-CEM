```python
## (C) 2020 UvA FACT AI
import cem
m = cem.Main(mode='PP', seed=121, type='MNIST')
# m.model_summary()

m.quant_eval(ids=[1,2,2953])
```

```python
def run_all(id, m, modes=['PP', 'PN'], gammas=[0,100], **args):
    """Run all models."""
    for mode in modes:
        m.set_mode(mode)
        for gamma in gammas:
            m.explain(id, gamma=gamma, **args)
```

```python
args = {'report':False, 'max_iter':1000}
```

```python
run_all(12, m, **args)
```
